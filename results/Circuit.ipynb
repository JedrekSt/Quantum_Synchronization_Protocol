<h1>Testing Shift Operator</h1>

```python
_0,_1 = [1,0],[0,1]
Experiment = ShiftExperiment_FlowControl(controlling_anc=2)

for i in range(8):
    string = Shift.DecToBin(i,3)
    tab = [_0,_0] + [_0 if string[j] == "0" else _1 for j in range(len(string))]
    res = Experiment.run(tab,show_circ =  True if i==0 else False)
    print(f"IniitialState : {string} , output : {res}")
```

<h1>4 nodes, 3 ancillas</h1>

```python
ancillas = 3
qubits = 2

_0 , _1 = [1,0],[0,1]
init_state_tab = [_1 if i%2 ==0 else _0 for i in range(ancillas)]  + [_0] * qubits
init_state_string = "".join(["1" if i%2 == 0 else "0" for i in range(ancillas)]) + "|" + ",".join(["H0"] * qubits)
model = SynchExperiment_FlowControl(ancillas,qubits)
res= model.run(init_state_tab,show_circ = True)

print(f"Result for the initial state {init_state_string}")
print("-"*10)
print(res)
print("-"*10)
```
