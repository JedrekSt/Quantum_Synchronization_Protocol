```python
import numpy as np
import matplotlib.pyplot as plt

n = 15
word = ("A" + "BBA" * (n-3)) * 1
fid = []
fs = 15

thetas = np.arange(0, np.pi/8, 0.01)

for th in range(len(thetas)):
    data = run_simulation(word, n, thetas[th])
    fid.append(data[-1, :].max())

fig, axes = plt.subplots(1, 2, figsize=(10, 4)) 

ax1 = axes[0]
data = run_simulation(word, n, np.pi/11)
im = ax1.imshow(data, origin="lower")
ax1.set_xlabel("position",fontsize = fs)
ax1.set_ylabel("time",fontsize = fs)
plt.colorbar(im, ax=ax1, fraction=0.05, pad=0.09)

ax2 = axes[1]
ax2.plot(thetas, fid, color="red", lw=3, marker="o",label = f"n = {n}")
ax2.set_xlabel(r"coin argument $\theta$",fontsize = fs)
ax2.set_ylabel(r"$\langle 1|\rho_{f}| 1\rangle$",fontsize = fs)

x_ticks = [0,np.pi/32,np.pi/16,np.pi * 3 / 32,np.pi/8]
x_labels = [r'$0$',r'$\frac{\pi}{32}$',r'$\frac{\pi}{16}$',r'$\frac{3\pi}{32}$',r'$\frac{\pi}{8}$']
plt.xticks(x_ticks, x_labels, fontsize=11)
plt.legend()

plt.tight_layout()
plt.savefig(os.path.join(os.path.curdir,"Images","QWalk.pdf"), format="pdf", bbox_inches="tight")
```
